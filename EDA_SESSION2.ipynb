# **Create dataframe using dictonary**

```python
name =['Avinash','Akash','Aadhaya']
age = [25,30,35]

pd.DataFrame(zip(name,age),columns = ['Names','Ages'])
```

```python
dictonary = ['mahathee:',12,'bhanu:',34,'cherry:',23]
pd.DataFrame(dictonary)
```

```python
dict1 = {'Names':['Avinash','Akash','Aadhaya'],
        'Ages':[12,14,15]}
pd.DataFrame(dict1)
#keys becomes columns
#Values becomes rows
#No need to provide column name separately
```

```python
dict2 = {'Names':'Avinash',
        'Age':23,
        'City':'Hyd'}
pd.DataFrame(dict2,index=[1])
```

```python
dict2 = {'Names':'Avinash',
        'Age':23,
        'City':'Hyd'}
pd.DataFrame(dict2,index=[1,2,3])
```
