**Data Tranformations**

- Generally data has 3 types of distributions

   - Normal distribution
    
   - Positive Skewed

   - Negative Skewed
    
- All the maths concepts developed by assumptions as Data follows Normal distributions

- Data skew happens because of many reasons,in that one reason is outliers

- Once we perform the outlier analysis then we can apply transformation methods,if the data does not follow normality

- We have some transformation methods mentioned below

   - Log Transformation
    
   - Exponential Transformation
    
   - Reciprocal Transformation
    
   - Square root Transformation
    
   - Power Transformation

**Step-1**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

**Step-2**

```python
exp_data = np.random.exponential(size=10000)
exp_data
```

**Step-3**

```python
plt.hist(exp_data,bins=50,label='Exponential')
plt.show()
```

```python
norm_data=np.random.normal(size=10000)
norm_data
plt.hist(norm_data,bins=50,label='Normal')
plt.show()
```

$Step-4$

**Log Transformation**

- Log Transformation means performing logarithm operations on original data

- It is an approach to convert into Normal distribution

- Log means natural logarithm base=e

```python
x=2
np.log(x)
```

```python
log_data=np.log(exp_data)
```

```python
plt.hist(log_data,bins=20,label='Logarithm')
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.subplot(1,2,1).hist(log_data,bins=50,label='Logarithm')
plt.legend()
plt.subplot(1,2,2).hist(exp_data,bins=50,label='Exponential')
plt.legend()
plt.show()
```

**Reciprocal Transformations**

- Reciprocal will fail when values are low

- For examples x=0 then 1/0 fail

```python
x=2
1/x
```

```python
1/exp_data
```

```python
1/log_data
```

```python
reciprocal_data=np.reciprocal(exp_data)
plt.hist(reciprocal_data,bins=20,label="Reciprocal")
plt.show()
```

```python
squared_data=np.sqrt(exp_data)
plt.hist(squared_data,bins=20,label="Square Root")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,3,1).hist(log_data,bins=50,label='Logarithm')
plt.legend()
plt.subplot(1,3,2).hist(exp_data,bins=20,label="Exponential")
plt.legend()
plt.subplot(1,3,3).hist(squared_data,bins=20,label="Square Root")
plt.legend()

plt.show()
```

**Power Transformers**

- Box-cox

- yeo-johnson

- It is under sklearn 

```python
from sklearn.preprocessing import PowerTransformer

pt=PowerTransformer(method='box-cox',standardize=True)

X_Transformed = pt.fit_transform(exp_data.reshape(-1, 1))

plt.hist(X_Transformed,bins=20,label='BOX-COX')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer

pt=PowerTransformer(method='box-cox',standardize=True)
data1=[[1,2],[3,5],[7,8]]
Y_Transformed = pt.fit_transform(data1)
print(Y_Transformed)
plt.hist(Y_Transformed)
```
